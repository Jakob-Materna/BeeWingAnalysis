### 5-AgeHistogram.ipynb

Creates an age histogram.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load dataset containing age information of live bees
data_path = "/mnt/c/Projects/Master/Data/Processed/LiveBees/7-AgeWings/LiveBeesAges.csv"
df = pd.read_csv(data_path)
```

```python
# Plot histogram of age distribution
plt.figure(figsize=(8, 5))
plt.hist(df["Age"], bins=20, edgecolor="black")

plt.xlabel("Age (days)", fontsize=12)
plt.ylabel("Frequency", fontsize=12)
plt.title("Distribution of Bee Ages", fontsize=14)

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
