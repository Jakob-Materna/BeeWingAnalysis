Visualizes model performance.

```python
import pandas as pd
import numpy as np
import os
import cv2
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
def evaluate_predictions(predictions_file, test_dir, actual_ages_file, title="Age Prediction Results", save_path=None):
    # Load predictions
    predictions_df = pd.read_csv(predictions_file)
    test_filenames = set(os.listdir(test_dir))
    df_filtered = predictions_df[predictions_df["Filename"].isin(test_filenames)].reset_index(drop=True)

    # Load actual ages
    actual_ages_df = pd.read_csv(actual_ages_file)

    # Rremove file extensions
    df_filtered["Filename"] = df_filtered["Filename"].str.replace(r'\.[a-zA-Z]{3,4}$', '', regex=True)
    actual_ages_df["Filename"] = actual_ages_df["Filename"].str.replace(r'\.[a-zA-Z]{3,4}$', '', regex=True)

    # Merge on 'Filename'
    merged_df = pd.merge(df_filtered, actual_ages_df, on="Filename")

    y_true = merged_df["Age"]
    y_pred = merged_df["PredictedAge"]

    # Calculate metrics
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)

    print(f"MAE: {mae:.2f}")
    print(f"RMSE: {rmse:.2f}")
    print(f"RÂ²: {r2:.2f}")

    plt.figure(figsize=(5, 4))
    
    # Ensure the plot starts at 0, 0
    min_val = min(min(y_true), min(y_pred)) * 0.95
    max_val = max(max(y_true), max(y_pred)) * 1.05
    # min_val = max(min_val, 0)  

    plt.xlim(0, max_val)
    plt.ylim(0, max_val)
    
    # Scatter plot with open black circles
    plt.scatter(y_true, y_pred, edgecolor='black', facecolor='none', alpha=0.8)
    plt.plot([min_val, max_val], [min_val, max_val], 'k--', linewidth=1)
    
    # Labels in serif font
    plt.xlabel("True Age (days)", fontsize=12, family='serif')
    plt.ylabel("Predicted Age (days)", fontsize=12, family='serif')
    
    # Remove grid and title
    plt.grid(False)
    plt.title("")
    plt.xticks(fontsize=10, family='serif')
    plt.yticks(fontsize=10, family='serif')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
            
    plt.show()
```

```python
def evaluate_weight_predictions(predictions_file, test_dir, actual_ages_file, title="Weight Prediction Results", save_path=None):
    # Load predictions
    predictions_df = pd.read_csv(predictions_file)
    test_filenames = set(os.listdir(test_dir))
    df_filtered = predictions_df[predictions_df["Filename"].isin(test_filenames)].reset_index(drop=True)

    # Load actual ages
    actual_ages_df = pd.read_csv(actual_ages_file)
    
    # Rremove file extensions
    df_filtered["Filename"] = df_filtered["Filename"].str.replace(r'\.[a-zA-Z]{3,4}$', '', regex=True)
    df_filtered['PredictedWeight'] = df_filtered['PredictedWeight'] # * 1000
    actual_ages_df["Filename"] = actual_ages_df["filename"].str.replace(r'\.[a-zA-Z]{3,4}$', '', regex=True)

    # Merge on 'Filename'
    merged_df = pd.merge(df_filtered, actual_ages_df, on="Filename")

    y_true = merged_df["weight"]
    y_pred = merged_df["PredictedWeight"]

    # Calculate metrics
    mae = mean_absolute_error(y_true, y_pred)
    rmse = np.sqrt(mean_squared_error(y_true, y_pred))
    r2 = r2_score(y_true, y_pred)

    # print(f"MAE: {mae:.2f}")
    # print(f"RMSE: {rmse:.2f}")
    # print(f"RÂ²: {r2:.2f}")

    residuals = y_true - y_pred
    plt.figure(figsize=(5, 4))
    
    # Get min and max values across both axes
    min_val = min(min(y_true), min(y_pred)) * 0.90
    max_val = max(max(y_true), max(y_pred)) * 1.05
    
    # Ensure min_val is not negative
    min_val = max(min_val, 0)
    
    # Set limits
    plt.xlim(min_val, max_val)
    plt.ylim(min_val, max_val)

    # Scatter plot with open black circles
    plt.scatter(y_true, y_pred, edgecolor='black', facecolor='none', alpha=0.8)
    plt.plot([min_val, max_val], [min_val, max_val], 'k--', linewidth=1)
    
    # Labels in serif font
    plt.xlabel("True Weight (g)", fontsize=12, family='serif')
    plt.ylabel("Predicted Age (g)", fontsize=12, family='serif')
    
    # Remove grid and title
    plt.grid(False)
    plt.title("")  # remove title if set
    
    # Ticks in serif font
    plt.xticks(fontsize=10, family='serif')
    plt.yticks(fontsize=10, family='serif')

    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
            
    plt.show()
```
