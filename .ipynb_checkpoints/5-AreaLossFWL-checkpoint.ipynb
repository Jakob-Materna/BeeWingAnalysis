### 5-AreaLossFWL.ipynb

Shows the relationship between bee age and forewing lobe area with a linear trend line.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import linregress
import statsmodels.api as sm
```

```python
# Load data
livebees_df = pd.read_csv("/mnt/c/Projects/Master/Data/Analysis/FWL-damage/LiveBees.csv")
ages_df = pd.read_csv("/mnt/c/Projects/Master/Data/Analysis/FWL-damage/ages.csv")
merged_df = pd.merge(livebees_df, ages_df, on="Filename")

# Filter for the FWL cell only
fwl_df = merged_df[merged_df["Cell"] == "FWL"]

# Drop rows with missing values in 'Age' or 'Area'
fwl_df = fwl_df.dropna(subset=["Age", "Area"])

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(fwl_df["Age"], fwl_df["Area"])

# Print regression result
print(f"Slope: {slope:.4f}")
print(f"P-value: {p_value:.4e}")
```

```python
# Linear regression
slope, intercept, r_value, p_value, std_err = linregress(fwl_df["Age"], fwl_df["Area"])
r_squared = r_value**2

# Create plot
plt.figure(figsize=(5, 4))

# Scatter plot with filled black points and white edges
plt.scatter(
    fwl_df["Age"], fwl_df["Area"],
    edgecolor='black', facecolor='none', alpha=0.8
)

# Regression line
x_vals = pd.Series(sorted(fwl_df["Age"]))
y_vals = intercept + slope * x_vals
plt.plot(x_vals, y_vals, 'k-', linewidth=1)

# Axis labels in serif font
plt.xlabel("Age (days)", fontsize=12, family='serif')
plt.ylabel("Forewing lobe area (mm)", fontsize=12, family='serif')

# Format ticks
plt.xticks(fontsize=10, family='serif')
plt.yticks(fontsize=10, family='serif')

# No grid, no title
plt.grid(False)
plt.title("")

plt.tight_layout()
# plt.savefig("4_age_vs_forewing_lobe_area.png", dpi=300)
plt.show()

# Print regression info
print(f"Slope: {slope:.4f} mm/day")
print(f"P-value: {p_value:.2e}")
print(f"R: {r_squared:.3f}")
print(f"n = {len(fwl_df)}")
```
